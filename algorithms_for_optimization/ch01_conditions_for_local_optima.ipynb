## 1.6 conditions for local optima

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.power(x,3) -3*x + 2
```

```python
def f_prime(x):
    return 3*np.power(x,2) -3
```

```python
def f_double_prime(x):
    return 6*x
```

```python
def f_triple_prime(x):
    return 6
```

```python
def taylor_1st(x,a):
    return f(a) + f_prime(a)*(x-a)
```

```python
# taylor expansion of f(x) up to second order at x=a
def taylor_2nd(x,a):
    return f(a) + f_prime(a)*(x-a) + 0.5*f_double_prime(a)*np.power(x-a,2)
```

```python
def taylor_3rd(x,a):
    return f(a) + f_prime(a)*(x-a) + 0.5*f_double_prime(a)*np.power(x-a,2) + 1/6*f_triple_prime(a)*np.power(x-a,3)
```

```python
x = np.linspace(-2,2,100)
y = f(x)
y_taylor_2nd = taylor_2nd(x,1)
y_taylor_3rd = taylor_3rd(x,1)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,label='f(x)')
plt.plot(x,taylor_1st(x,1),label='1st order Taylor', linestyle='dashed')
plt.plot(x,y_taylor_2nd,label='2nd order Taylor')
plt.plot(x,y_taylor_3rd,label='3rd order Taylor', linestyle='dotted')
plt.scatter(1,f(1),color='red')
plt.legend()
plt.show()
```

### 1.6.2 multivariate

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(x):
    return np.power(1-x[0],2) + 5*np.power(x[1]-np.power(x[0],2),2)
```

```python
x = np.linspace(-2, 2, 400)
y = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x, y)
Z = f([X, Y])
```

```python
plt.figure(figsize=(8,6))
cp = plt.contour(X, Y, Z, levels=np.logspace(-1, 3, 20), cmap='viridis')
plt.colorbar(cp)
plt.title('Contour plot of the function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Contour Plot')
plt.show()
```
